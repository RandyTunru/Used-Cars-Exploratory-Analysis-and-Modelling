# Data Preprocessing

We will remove features that aren't descriptive 

```python
excluded_columns = ["id", "region_url", "url", "image_url"]
```

```python
main_dataset = vehicles_dataset.loc[ : , [col for col in vehicles_dataset if col not in excluded_columns]]
main_dataset
```

The column "county" has all of it's values missing, we decided to drop all of it's values

```python
main_dataset = main_dataset.drop(columns=['county'])
```

```python
main_dataset.info()
```

Let's also differentiate the columns based on it's type

```python
numerical_features = [col for col in main_dataset.columns
                      if  (main_dataset[col].dtype != "object")
                      ]

categorical_features = [col for col in main_dataset.columns
                      if  (main_dataset[col].dtype == "object")
                      ]

print("Numerical Features: ", numerical_features)
print("Categorical Features: ", categorical_features)
```

## Outlier Detection

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Let's do outlier detection on our numerical features

```python
for i, col in enumerate(numerical_features):
    print(f"Plotting {col}")
    sns.boxplot(x=main_dataset[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
    print(main_dataset[col].describe())
    print("\n")
```

These values are quite hard to handle it's outliers let's tackle this 1 by 1

### Price

```python
sns.boxplot(x=main_dataset["price"])
plt.title(f"Boxplot of price")
plt.xlabel('price')
plt.show()
print(main_dataset["price"].describe())
```

A few things that can be pointed out is 
    1. There are cars with extremely high prices (over around 0.9 * 1e9 = 500,000,000)
    2. There are cars with price 0

```python
main_dataset.where(main_dataset["price"] > 9e8).dropna(subset=["price"]).sort_values(by="price", ascending=False)
```

After inspecting the cars manually, we suspect that this might be misinputs

Let's assume (by domain knowledge) that for used cars the price would realistically be atleast 10,000 and most expensive would be 500,000

```python
main_dataset = main_dataset[(main_dataset["price"] > 10000) & (main_dataset["price"] < 500000)]
```

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x=main_dataset["price"])
plt.title(f"Boxplot of price")
# ax.set_xticks(np.arange(0, 1100000, 50000))
plt.xlabel('price')
plt.show()
print(main_dataset["price"].describe())
```

```python
main_dataset.sort_values(by="price", ascending=False).head(10)
```

Doing manual validation on this, the price on this  makes sense hence we assume there's already minimal outliers left
