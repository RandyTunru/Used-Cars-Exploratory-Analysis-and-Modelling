### Year

Handling this can be tricky as we can't immediately assume it would be a common value among the dataset (mean, median). 
One way i thought of doing is using regex on the description to find a possible year value. Another way i thought to do is to look at the same car model and use the median value for it (for that specific car model, which year of production is most likely to be sold based on our data). If it's still missing then we can just drop it

```python
main_dataset.description.sample(20, random_state=42)
```

```python
import re
```

```python
main_dataset['year_from_desc'] = (
    main_dataset['description']
      .fillna('')
      .str.extract(r'\b(19\d{2}|20\d{2})\b', expand=False)
      .astype('float') # temporarily float so NaNs are represented
      .fillna(np.nan)  # convert to NaN for missing values
)
```

```python
# 1. Define a function to grab the first mode (if it exists)
def most_common_year(series):
    modes = series.mode()
    return int(modes.iloc[0]) if not modes.empty else np.nan

# 2. Group by model and apply
model_mode_year = (
    main_dataset
     .groupby('model')['year']
     .agg(most_common_year)
     .dropna()         # drop models that had no valid years at all
     .astype(int)
     .to_dict()
)
```

```python
# 1. Try regex‐extracted year
mask_missing = main_dataset['year'].isna()
main_dataset.loc[mask_missing, 'year'] = main_dataset.loc[mask_missing, 'year_from_desc']

# 2. Still missing?  Look up by model mode
mask_still_missing = main_dataset['year'].isna()
main_dataset.loc[mask_still_missing, 'year'] = (
    main_dataset.loc[mask_still_missing, 'model']
                .map(model_mode_year)
)

# 3. Drop any rows where year is STILL missing
before = len(main_dataset)
main_dataset = main_dataset.dropna(subset=['year'])
dropped = before - len(main_dataset)

# 4. Cast to int
main_dataset['year'] = main_dataset['year'].astype(int)
print(f"Dropped {dropped} rows where year was still missing")
main_dataset.drop(columns=['year_from_desc'], inplace=True)
```

### Manufacturer

We can use a similar approach we did with the "year" column. 
- Find the brand mentioned in the description if the description only mentions one manufacturer name. 
- If not found yet, we'll look at other data of the same model name and copy it's manufacturer. 
- If still not found we'll impute as "Unknown"

```python
unique_manufacturers = main_dataset['manufacturer'].unique()
# Strip + lowercase + drop missing
main_dataset['manufacturer'] = (
    main_dataset['manufacturer']
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({'nan': pd.NA})
)
unique_manufacturers = set(main_dataset['manufacturer'].dropna().unique())
```

```python
print(unique_manufacturers)
```

```python
model_manufacturer_pairs = (
    main_dataset
    .dropna(subset=['manufacturer'])
    .groupby('model')['manufacturer']
    .agg(lambda s: s.mode().iat[0] if not s.mode().empty else pd.NA)
    .dropna()
    .to_dict()
)
```

```python
# Build a regex like r'\b(ford|toyota|honda)\b'
pattern = r'\b(' + '|'.join(re.escape(m) for m in unique_manufacturers) + r')\b'

# Extract all occurrences of manufacturers from the description 
main_dataset['found_mfrs'] = (
    main_dataset['description']
    .fillna('')
    .str.lower()
    .str.findall(pattern)
)
```

```python
copy_main_dataset = main_dataset.copy()
```

```python
main_dataset['unique_found_mfrs'] = main_dataset['found_mfrs'].apply(lambda x: tuple(x))
```

```python
for i, row in main_dataset.sample(10, random_state=42).iterrows():
    print(f"Row {i}: {row['description']}")
    print(f"Found manufacturers: {row['found_mfrs']}")
    print()

# Convert lists to sets to remove duplicate mentions (e.g., if "ford" appears multiple times in the description)

for i, row in main_dataset.sample(10, random_state=42).iterrows():
    print(f"Row {i}: {row['description']}")
    print(f"Unique Found manufacturers: {row['unique_found_mfrs']}")
    print()
```

```python
single_mfr = main_dataset.loc[((main_dataset['unique_found_mfrs'].str.len() == 1) & main_dataset['manufacturer'].isna())]
```

```python
main_dataset.loc[single_mfr.index, 'manufacturer'] = main_dataset.loc[single_mfr.index, 'unique_found_mfrs'].apply(lambda x: x[0])
```

```python
# If manufacturer is still missing, try to extract from same models
mask = main_dataset['manufacturer'].isna() & main_dataset['model'].notna()
main_dataset.loc[mask, 'manufacturer'] = (
    main_dataset.loc[mask, 'model']
                .map(model_manufacturer_pairs)
)

# Drop rows where manufacturer is still missing
main_dataset['manufacturer'] = main_dataset['manufacturer'].fillna("Unknown")

unknown_manufacturer = main_dataset[main_dataset.manufacturer == 'Unknown']

print(f"{len(unknown_manufacturer)} rows where manufacturer is Unknown")
main_dataset.drop(columns=['found_mfrs', 'unique_found_mfrs'], inplace=True)
```
